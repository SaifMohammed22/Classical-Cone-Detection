```python
import cv2
import numpy as np
from time import sleep
```

```python
video = cv2.VideoCapture('Car.mp4')
subtracao = cv2.createBackgroundSubtractorKNN(history=100, dist2Threshold=4000)
delay = 10
```

```python
for i in range(10):  # Allow 10 frames to initialize
    ret, frame = video.read()
    if not ret:
        break 
    subtracao.apply(frame)
```

```python
# Main loop
while True:
    ret, frame = video.read()
    if not ret:
        break
    
    # Set tempo delay
    sleep(1 / delay)
    
    # Define Regions of Interest (ROIs)
    roi_1 = frame[300:800, 1400:1900]  # Right region of interest
    roi_2 = frame[300:800, 0:500]      # Left region of interest
    
    # Convert ROIs to HSV color space
    hsv_1 = cv2.cvtColor(roi_1, cv2.COLOR_BGR2HSV)
    hsv_2 = cv2.cvtColor(roi_2, cv2.COLOR_BGR2HSV)

    # Define color ranges for filtering
    lower_white = np.array([0, 0, 180], dtype=np.uint8)
    upper_white = np.array([180, 40, 255], dtype=np.uint8)

    # lower_red1 = np.array([0, 70, 50], dtype=np.uint8)
    # upper_red1 = np.array([10, 255, 255], dtype=np.uint8)
    # lower_red2 = np.array([170, 70, 50], dtype=np.uint8)
    # upper_red2 = np.array([180, 255, 255], dtype=np.uint8)

    # Generate masks white colors
    # red_mask_1 = cv2.inRange(hsv_1, lower_red1, upper_red1) + cv2.inRange(hsv_1, lower_red2, upper_red2)
    # red_mask_2 = cv2.inRange(hsv_2, lower_red1, upper_red1) + cv2.inRange(hsv_2, lower_red2, upper_red2)

    white_lane_mask_1 = cv2.inRange(hsv_1, lower_white, upper_white)
    white_lane_mask_2 = cv2.inRange(hsv_2, lower_white, upper_white)

    # Background subtraction and morphology operations for ROI 1
    mask_1 = subtracao.apply(roi_1)
    # Filter out white regions in ROI 1
    mask_1 = cv2.bitwise_and(mask_1, mask_1, mask=cv2.bitwise_not(white_lane_mask_1))
    # mask_1 = cv2.bitwise_and(mask_1, mask_1, mask=cv2.bitwise_not(red_mask_1))
    _, mask_1 = cv2.threshold(mask_1, 254, 255, cv2.THRESH_BINARY)

    # Find and draw contours for ROI 1
    contours_1, _ = cv2.findContours(mask_1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    for cnt_1 in contours_1:
        area_1 = cv2.contourArea(cnt_1)
        if area_1 > 800:
            x, y, w, h = cv2.boundingRect(cnt_1)
            cv2.rectangle(roi_1, (x, y), (x+w, y+h), (0, 255, 0), 3)

    # Background subtraction and morphology operations for ROI 2
    mask_2 = subtracao.apply(roi_2)
    # Filter out white regions in ROI 2
    mask_2 = cv2.bitwise_and(mask_2, mask_2, mask=cv2.bitwise_not(white_lane_mask_2))
    # mask_2 = cv2.bitwise_and(mask_2, mask_2, mask=cv2.bitwise_not(red_mask_2))
    _, mask_2 = cv2.threshold(mask_2, 254, 255, cv2.THRESH_BINARY)

    # Find and draw contours for ROI 2
    contours_2, _ = cv2.findContours(mask_2, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt_2 in contours_2:
        area_2 = cv2.contourArea(cnt_2)
        if area_2 > 800:
            x, y, w, h = cv2.boundingRect(cnt_2)
            cv2.rectangle(roi_2, (x, y), (x+w, y+h), (255, 0, 0), 3)

    # Display results
    cv2.imshow('Original Video', frame)

    # For debugging
    
    # cv2.imshow('ROI "Right side"', mask_1)
    # cv2.imshow('ROI Video "Left side"', mask_2)
    # cv2.imshow('ROI 1', roi_1)
    # cv2.imshow('ROI 2', roi_2)
    # cv2.imshow('HSV Image 1', hsv_2)
    # cv2.imshow('White Mask 1', white_lane_mask_2)
    # cv2.imshow('Red Mask 1', hsv_1)
    # cv2.imshow('White Mask 1', white_lane_mask_1)


    # Exit loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
